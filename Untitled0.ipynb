````python
backend_code = r'''
# backend_app.py
from fastapi import FastAPI, HTTPException, Depends
from fastapi.security import OAuth2PasswordRequestForm
from pydantic import BaseModel
from sqlalchemy import create_engine, Column, Integer, String, Float, JSON
from sqlalchemy.orm import sessionmaker, declarative_base
from passlib.context import CryptContext
import jwt, time, os, json
from typing import List
from fastapi.middleware.cors import CORSMiddleware
from tensorflow.keras.models import load_model
import numpy as np

SECRET = os.environ.get("SECRET","supersecretkey")
ALGO = "HS256"

Base = declarative_base()
DB_PATH = os.environ.get("DB_PATH","sqlite:///./app.db")
engine = create_engine(DB_PATH, connect_args={"check_same_thread": False})
SessionLocal = sessionmaker(bind=engine)

pwd_context = CryptContext(schemes=["bcrypt"], deprecated="auto")

class User(Base):
    __tablename__ = "users"
    id = Column(Integer, primary_key=True, index=True)
    username = Column(String, unique=True, index=True)
    hashed_password = Column(String)
    email = Column(String, unique=True, index=True)

class Portfolio(Base):
    __tablename__ = "portfolios"
    id = Column(Integer, primary_key=True, index=True)
    user_id = Column(Integer, index=True)
    holdings = Column(JSON)  # {"AAPL": {"qty": 10, "avg_price": 150.0}, ...}
    cash = Column(Float, default=100000.0)

Base.metadata.create_all(bind=engine)

app = FastAPI(title="Trading Backend (demo)")
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

class Signup(BaseModel):
    username: str
    password: str
    email: str

class TradeRequest(BaseModel):
    symbol: str
    qty: int
    side: str  # "buy" or "sell"
    price: float = None

def get_db():
    db = SessionLocal()
    try:
        yield db
    finally:
        db.close()

def create_token(user_id: int):
    payload = {"user_id": user_id, "exp": time.time() + 60*60*24}
    return jwt.encode(payload, SECRET, algorithm=ALGO)

def verify_token(token: str):
    try:
        payload = jwt.decode(token, SECRET, algorithms=[ALGO])
        return payload.get("user_id")
    except Exception:
        return None

def get_user_by_username(db, username):
    return db.query(User).filter(User.username==username).first()

@app.post("/signup")
def signup(data: Signup, db=Depends(get_db)):
    if get_user_by_username(db, data.username):
        raise HTTPException(400, "Username exists")
    # Truncate password to 72 bytes before hashing, as required by bcrypt
    hashed = pwd_context.hash(data.password.encode('utf-8')[:72])
    u = User(username=data.username, hashed_password=hashed, email=data.email)
    db.add(u); db.commit(); db.refresh(u)
    p = Portfolio(user_id=u.id, holdings=json.dumps({}), cash=100000.0)
    db.add(p); db.commit()
    return {"msg":"user created"}

@app.post("/login")
def login(form: OAuth2PasswordRequestForm = Depends(), db=Depends(get_db)):
    user = get_user_by_username(db, form.username)
    # bcrypt.verify automatically truncates passwords internally for comparison
    if not user or not pwd_context.verify(form.password, user.hashed_password):
        raise HTTPException(401, "Invalid credentials")
    token = create_token(user.id)
    return {"access_token": token, "token_type":"bearer"}

from fastapi import Header
def get_current_user(authorization: str = Header(None), db=Depends(get_db)):
    if not authorization: raise HTTPException(401,"Missing token")
    token = authorization.split(" ")[1] if " " in authorization else authorization
    uid = verify_token(token)
    if not uid: raise HTTPException(401,"Invalid token")
    user = db.query(User).filter(User.id==uid).first()
    return user

@app.get("/me")
def me(user = Depends(get_current_user)):
    return {"username": user.username, "email": user.email}

@app.get("/portfolio")
def get_portfolio(user = Depends(get_current_user), db=Depends(get_db)):
    p = db.query(Portfolio).filter(Portfolio.user_id==user.id).first()
    return {"holdings": p.holdings or {}, "cash": p.cash}

@app.post("/trade")
def trade(req: TradeRequest, user=Depends(get_current_user), db=Depends(get_db)):
    p = db.query(Portfolio).filter(Portfolio.user_id==user.id).first()
    holdings = p.holdings or {}
    price = req.price if req.price else 100.0
    cost = price * req.qty
    if req.side.lower()=="buy":
        if p.cash < cost:
            raise HTTPException(400, "Insufficient cash")
        p.cash -= cost
        h = holdings.get(req.symbol, {"qty":0,"avg_price":0.0})
        new_qty = h["qty"] + req.qty
        new_avg = (h["avg_price"]*h["qty"] + price*req.qty) / new_qty if new_qty>0 else 0.0
        holdings[req.symbol] = {"qty": new_qty, "avg_price": new_avg}
    else:
        h = holdings.get(req.symbol)
        if not h or h["qty"] < req.qty:
            raise HTTPException(400, "Insufficient shares")
        h["qty"] -= req.qty
        proceeds = price * req.qty
        p.cash += proceeds
        if h["qty"]==0:
            holdings.pop(req.symbol, None)
        else:
            holdings[req.symbol] = h
    p.holdings = holdings
    db.add(p); db.commit()
    return {"status":"ok", "holdings": holdings, "cash": p.cash}

# Model serving: optional - loads a Keras model file if exists
MODEL_PATH = os.environ.get("MODEL_PATH", "./bilstm.h5")
model = None
def load_prediction_model():
    global model
    try:
        model = load_model(MODEL_PATH)
        print("Model loaded:", MODEL_PATH)
    except Exception as e:
        print("No model available or failed to load:", e)

@app.on_event("startup")
def startup_event():
    load_prediction_model()

@app.post("/predict/{symbol}")
def predict(symbol: str, window: List[float]):
    # window: list of last N prices (length must match model input window)
    if model is None:
        raise HTTPException(500, "Model not available")
    arr = np.array(window)[None,:,None]  # shape (1, window, 1)
    pred = model.predict(arr)[0,0]
    return {"symbol": symbol, "prediction": float(pred)}
'''
open("backend_app.py","w").write(backend_code)
print("backend_app.py created")
````
